```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def apply_negative_transformation(image_array):
    neg_image_array = 255 - image_array
    return neg_image_array.astype(np.uint8)
```

```python
base_path = "/content/drive/MyDrive/ImageProcessingHomework/"
file_names = ["coins.ascii.pgm", "fruit.pgm", "saturn.ascii.pgm"]
```

```python
for file_name in file_names:
    if file_name == ".DS_Store":
        continue  # .DS_Store dosyasını atla

    image_path = base_path + file_name

    # Load the image
    input_image = Image.open(image_path).convert('L')
    image_array = np.array(input_image)

    # Apply negative transformation
    neg_image_array = apply_negative_transformation(image_array)

    # Display the original and adjusted images
    print(f"Original Image: {file_name}")
    plt.imshow(image_array, cmap='gray')
    plt.axis('off')
    plt.show()

    print(f"Negative Transformed Image: {file_name}")
    plt.imshow(neg_image_array, cmap='gray')
    plt.axis('off')
    plt.show()
```
